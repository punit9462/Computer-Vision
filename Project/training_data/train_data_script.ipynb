```python
from itertools import islice

def take(n, iterable):
    "Return first n items of the iterable as a list"
    return list(islice(iterable, n))
```

```python
with open('1.1') as f:
    lines = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
lines = [x.strip() for x in lines]
lines = ['dummy'] + lines
start_dictionary = {"are": 1, "can": 7, "could":8, "do": 9, "does": 11, "has": 13, "how": 14, "is": 18, "was": 33, "what": 34, "where": 60, "which": 62, "who": 63, "why": 64}
end_dictionary = {"are": 6, "can": 7, "could":8, "do": 10, "does": 12, "has": 13, "how": 17, "is": 31, "was": 33, "what": 59, "where": 61, "which": 62, "who": 63, "why": 65}

new_data_question = []
non_standard = 0
for data_question in data_questions:
    question = data_question['question'].lower()
#     question = "can you run?"
    question_words = question.split()
    question_first_word = question_words[0]
    if question_first_word in start_dictionary:
        start_index = start_dictionary[question_first_word]
        end_index = end_dictionary[question_first_word]
        actual_question_type = ""
        max_count = 0
        for outer_index in range(start_index, end_index + 1):
            question_type_line = lines[outer_index]
            question_type = question_type_line.split()
            count = 0
            index = 0
            for question_word in question_words:
                if(question_word == question_type[index]):
                    count = count + 1
                    index = index + 1
                    if(index == len(question_type)):
                        break
                else:
                    break
            if(count > max_count):
                max_count = count
                actual_question_type = question_type_line
        data_question['question_type'] = actual_question_type
    else:
        print('first word is not in dictionary ' + question + ' ' + str(data_question['question_id']))
        non_standard = non_standard + 1
        data_question['question_type'] = "dummy"
    new_data_question.append(data_question)
print ('nstd ' + str(non_standard))
print ('total' + str(len(data_questions)))

# f = open('new_data_question.txt','w')
# f.write(str(new_data_question))
# f.close()
```
